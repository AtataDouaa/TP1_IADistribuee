```python
tokenizer = tiktoken.get_encoding("o200k_base")
```

```python
prompt = "En tant qu'expert, donne moi la definition d'un agent"
```

```python
tokens = tokenizer.encode(prompt)
for token in tokens:
    t=tokenizer.decode_single_token_bytes(token).decode("utf-8")
    print(t, end="|") 
```
